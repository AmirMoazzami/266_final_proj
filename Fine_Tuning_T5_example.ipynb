```python
import pandas as pd
import numpy as np
from simplet5 import SimpleT5
```

```python
df1=pd.read_csv('Dataset_excel_1.csv',delimiter='\t',index_col=0)
```

```python
df2=pd.read_csv('Dataset_excel_2.csv',delimiter='\t',index_col=0)
```

```python
df3=pd.read_csv('Dataset_excel_3.csv',delimiter='\t',index_col=0)
```

```python
df1=df1.iloc[1:]
```

```python
df2=df2.iloc[1:]
```

```python
df3=df3.iloc[1:]
```

```python
frames = [df1, df2,df3]
```

```python
df = pd.concat(frames,ignore_index=True)
```

```python
df = df.drop('adress', axis=1)
```

```python
df
```

```python
df_train=df.iloc[0:35500]
df_test=df.iloc[35500:]
```

```python
df_train = df_train.rename(columns={"correct":"target_text", "incorrect":"source_text"})
```

```python
df_test = df_test.rename(columns={"correct":"target_text", "incorrect":"source_text"})
```

```python
model=SimpleT5()
model.from_pretrained(model_type="mt5",model_name='Ahmad/parsT5-base')
```

```python
df_train
```

```python
model.train(train_df=df_train,
            eval_df=df_test, 
            source_max_token_len=40, 
            target_max_token_len=40, 
            batch_size=8, max_epochs=5, use_gpu=True)
```
